```python
import pandas as pd
import numpy as np
import spacy
import re
DIR = 'OtherLists'
path = DIR + '/Got3.txt'
got = open(path).read()

print(got[:200])

got = "363.".join(got.split("...")) 
got_dot = got.split(".")
    
sents = []
for i, sent in enumerate(got_dot):
    sent = " ".join(sent.split("\n"))
    sent = sent.strip()
    if sent.startswith('” '):
        sents[i-1] += '”'
        sent = sent[2:]
        sent = sent.strip()
    if not sent.endswith('”'):
        sent += "."
    if '363' in sent:
        sent = sent.replace('363', '..')
    sents.append(sent)

def add_dialogs(sents):
    joined_sents = []
    for i, sent in enumerate(sents):
        if not sent.startswith('“') and sent.endswith('”'):
            joined_sents[i-1] += " "+sent
            joined_sents.append("369")
        elif sent[0].islower() and not sent.endswith('”'):
            joined_sents[i-1] += " "+sent
            joined_sents.append("369")
        else:
            joined_sents.append(sent)
    return joined_sents

for i in range(20):
    sents = add_dialogs(sents)
    sents = ["".join(sent.split('369')) for sent in sents if sent is not '369']

sents2 = []
[(sents2.extend(sent.split('  '))) for sent in sents]

sents = sents2

# Author typo fix
s2 = []
for i, sent in enumerate(sents):
    if sent.startswith('And your hair, your golden hair...'):
        s2[i-1] += " "+sent+'”'
    else:
        s2.append(sent)
    
def connect_dialogs(lines):
    sents3 = []
    for i, sent in enumerate(lines):
        cco = sent.count('“')
        cce = sent.count('”')
        if cco is 1 and cce is 0:
            sents3.append(sent)
        elif cce is 1 and cco is 0:
            sents3[i-1] += " "+sent
            sents3.append('345')
        elif cce > 1 or cco > 1:
            sents3.append(sent)
        else:
            sents3.append(sent)
    sents3 = [sent.strip() for sent in sents3 if sent is not '345']
    return sents3

sents3 = s2
for i in range(100):
    sents3 = connect_dialogs(sents3)

sents4 = sents3

sents5 = []
for i, sent in enumerate(sents4):
    cco = sent.count('“')
    cce = sent.count('”')
    if cce > 1 or cco > 1:
        t = 0
        if '.”' in sent:
            t+=1
            idx = sent.index('.”') + 2
            sent3, sent4 = sent[:idx], sent[idx:]
            sents5.append(sent3)
            if sent4:
                sents5.append(sent4)

        if '?” ' in sent:
            t+=1
            idx = sent.index('?” ')
            split_on = sent.find('.',idx) + 1
            for snt in [sent[:split_on], sent[split_on:]]:
                if not snt.endswith('”') and not snt.endswith('.'):
                    snt += '.'
                if len(snt) > 1:
                    sents5.append(snt)
        if t is 0:
            sents5.append(sent)
    else:
        sents5.append(sent)
sents5 = [s.strip() for s in sents5]

sents6 = []
for i, sent in enumerate(sents5):
    cco = sent.count('“')
    cce = sent.count('”')
    if cce is 0 and cco is 0 and sent.count('? ') > 0:
        split_sent = sent.split('? ')
        if len(split_sent) is 2:
            s1, s2 = split_sent
            s1 += '?'
            sents6.append(s1)
            sents6.append(s2)
        else:
            print("\n\n\nUnique split alert\n\n\n", sent,'\n\n\n')
    else:
        sents6.append(sent)
        
sents7 = []
for i, sent in enumerate(sents6):
    cco = sent.count('“')
    cce = sent.count('”')
    if (cce > 0 or cco > 0) and sent.count(". ") > 1:
        cco_i = [m.start() for m in re.finditer('“', sent)]
        cce_i = [m.start() for m in re.finditer('”', sent)]
        dot_i = [m.start() for m in re.finditer('[.]', sent)]
        quotes = list(zip(cco_i, cce_i))
        q2 = [range(r[0],r[1]) for r in quotes]
        good_dots = [0]
        for i in dot_i:
            [good_dots.append(i+1) for each in q2 if i not in each]
        parts = [sent[i:j] for i,j in zip(good_dots, good_dots[1:]+[None])]
        for part in parts:
            if len(part) > 1:
                part = part.strip()
                sents7.append(part)
    else:
        sents7.append(sent)
        

sents8 = []
for i, sent in enumerate(sents7):
    if '?”' in sent:
        idx = sent.index('?”')
        if idx+3 < len(sent):
            if sent[sent.index('?”')+3].isupper():
                sent1, sent2 = sent.split('?”')
                sent1 = sent1+'?”'
                sents8.append(sent1)
                sents8.append(sent2)
            else:
                sents8.append(sent)
        else:
            sents8.append(sent)
    else:
        sents8.append(sent)

for i, sent in enumerate(sents8):
    print(i, sent)
```
